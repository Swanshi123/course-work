```python
import pandas as pd
```

```python
df_price=pd.read_excel(r'./分省房价数据.xlsx')
df_price
```

```python
df=pd.read_excel(r'./CGSS2021.xlsx')
```

```python
df.info()
```

```python
# Perform a left join
df['省份'] = df['省份'].astype(str)
df = pd.merge(df, df_price, on='省份', how='left')
df
```

```python
df.info()
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.preprocessing import MinMaxScaler

# Step 1: Load and clean the data
selected_columns = ['省份', '最高教育程度', '所住房屋面积', '个人2020年收入', '2021年房价','希望有几个孩子']
df_7001 = df[selected_columns].copy()

# Step 2: Handle non-numeric and missing values
df_7001['所住房屋面积'] = pd.to_numeric(df_7001['所住房屋面积'], errors='coerce')#
df_7001['个人2020年收入'] = pd.to_numeric(df_7001['个人2020年收入'], errors='coerce')
df_7001['希望有几个孩子'] = pd.to_numeric(df_7001['希望有几个孩子'], errors='coerce')
#df['2021年房价'] = pd.to_numeric(df['2021年房价'], errors='coerce')
#This function can handle non-numeric data, convert it to numeric types, and return NaN (Not a Number) if the conversion fails.
```

```python
# Drop rows with missing or invalid data after conversion
df_7001.dropna(subset=['所住房屋面积', '个人2020年收入', '希望有几个孩子','2021年房价'], inplace=True)

# Step 3: Convert categorical variables to numeric，.astype('category').cat.codes，Education classification can be further processed
df_7001['省份'] = df_7001['省份'].astype('category').cat.codes
df_7001['最高教育程度'] = df_7001['最高教育程度'].astype('category').cat.codes

# Step 4: Normalize numerical columns (house area, personal income)
scaler = MinMaxScaler()
df_7001[['所住房屋面积', '个人2020年收入','2021年房价']] = scaler.fit_transform(df_7001[['所住房屋面积', '个人2020年收入','2021年房价']])

# Step 5: Convert target variable to binary (0 = fewer than 2 children, 1 = 2 or more children)
df_7001['高生育意愿'] = df_7001['希望有几个孩子'].apply(lambda x: 1 if x >= 2 else 0)

# Step 6: Define features (X) and target (y)
X = df_7001[['省份', '最高教育程度', '所住房屋面积', '个人2020年收入','2021年房价']]
y = df_7001['高生育意愿']

# Add a constant term to the model (intercept)
X = sm.add_constant(X)

# Step 7: Fit the logistic regression model using statsmodels
model = sm.Logit(y, X)
result = model.fit()
```

```python
# Print the full model summary
print("\nFull Model Summary:")
print(result.summary())
```

Detailed Interpretation:

	1.	Intercept (const = 1.2942):
	•	The positive intercept suggests a baseline tendency towards high fertility intention when all predictors are zero.
	•	The p-value (0.000) indicates that the intercept is highly significant.
	2.	Province (省份):
	•	Coefficient: -0.0049, Odds Ratio ≈ e^{-0.0049} \approx 0.995.
	•	Interpretation: The effect of province on fertility intention is very small and not statistically significant (p-value = 0.425). This suggests that regional differences, as captured by the province codes, do not have a significant effect on fertility intention in this model.
	3.	Education Level (最高教育程度):
	•	Coefficient: 0.0045, Odds Ratio ≈ e^{0.0045} \approx 1.005.
	•	Interpretation: The effect of education level on fertility intention is almost negligible and not statistically significant (p-value = 0.641). This implies that, in your dataset, education level does not significantly influence fertility intentions.
	4.	House Area (所住房屋面积):
	•	Coefficient: 4.8154, Odds Ratio ≈ e^{4.8154} \approx 123.6.
	•	Interpretation: A strong positive effect, indicating that larger house areas are strongly associated with higher fertility intention. The odds of having a high fertility intention increase dramatically as the house area increases.
	•	The p-value (0.000) indicates that this variable is highly significant.
	5.	Personal Income in 2020 (个人2020年收入):
	•	Coefficient: -1.1625, Odds Ratio ≈ e^{-1.1625} \approx 0.31.
	•	Interpretation: There is a negative association between personal income and fertility intention, but this effect is not statistically significant (p-value = 0.166). Higher income tends to reduce the likelihood of high fertility intention, though the evidence is not strong enough to be conclusive.
	6.	Housing Price in 2021 (2021年房价):
	•	Coefficient: -1.0659, Odds Ratio ≈ e^{-1.0659} \approx 0.34.
	•	Interpretation: There is a significant negative effect of housing prices on fertility intention. Higher housing prices are associated with a decrease in fertility intentions.
	•	The p-value (0.000) indicates that this variable is highly significant, suggesting that housing prices are a strong deterrent to having more children.

Key Insights:

	•	Significant Factors:
	•	House Area: Strong positive effect on fertility intention, indicating that families with larger houses are more likely to want more children.
	•	Housing Prices: Strong negative effect, suggesting that high housing prices may discourage people from having more children.
	•	Non-significant Factors:
	•	Province, Education Level, and Personal Income did not show significant effects on fertility intention in this model.

Recommendations:

	•	Policy Implication: If housing prices are indeed a strong deterrent, policies aimed at reducing housing costs or providing housing support could positively influence fertility intentions.
	•	Further Analysis: It might be worthwhile to explore interactions between income, education, and housing prices, as these factors might have a combined effect that isn’t captured in the current model.
