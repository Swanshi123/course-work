```python
import numpy as np
df=df_new.copy()
```

```python
import statsmodels.api as sm


# 假设我们对'feature1', 'feature2', 'feature3'这三列取对数
df['GDP'] = np.log1p(df['GDP'])
#df_new['student-teacher ratio'] = np.log1p(df_new['student-teacher ratio'])
df['library collection'] = np.log1p(df['library collection'])
#df_new['University Acceptance Rate']== np.log1p(df_new['University Acceptance Rate'])



X = df[['GDP', 'student-teacher ratio', 'library collection']]# 特征列
X1= df[['GDP', 'student-teacher ratio']]
X2= df[['GDP']]


#X = df_new[['GDP', 'student-teacher ratio']]  # 特征列
y = df['University Acceptance Rate']  # 目标变量列

# 为特征数据添加一列常数项，以便进行线性回归
X = sm.add_constant(X)

# 创建线性回归模型
model = sm.OLS(y, X).fit()
model1 = sm.OLS(y, sm.add_constant(X1)).fit()
model2 = sm.OLS(y, sm.add_constant(X2)).fit()

# 打印模型的摘要，其中包括参数估计、t值、p值等统计信息
#print(model.summary())
```
